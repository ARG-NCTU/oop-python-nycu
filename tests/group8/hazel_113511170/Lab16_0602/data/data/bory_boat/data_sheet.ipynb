## Reading CSVs

read csv into
1. wamvs: dictionary of GeoDataFrame
2. trajs: dictionary of Trajectory

```python
boats = ['wamv1', 'wamv2', 'wamv3', 'wamv4', 'wamv5']
wamvs = {}
trajs = {}

for boat in boats:
    #df = pd.read_csv( 'data/data/bory_boat/'+ boat + '_gps.csv', delimiter=',')
    df = pd.read_csv(  boat + '_gps.csv', delimiter=',')
    df['t'] = pd.to_datetime(df['t'], unit='ns')
    wamv = df[['X', 'Y', 'trajectory_id', 't']]
    wamv['geometry'] = df.apply(lambda row: Point(row['X'], row['Y']), axis=1)

    gdf = gpd.GeoDataFrame(wamv, geometry='geometry', crs='EPSG:4326')
    gdf = gdf.to_crs(epsg=32649) # Convert to UTM zone 49N for Taiwan

    trajs[boat] = mpd.Trajectory(gdf, 'trajectory_id', t='t')
    wamvs[boat] = gdf
```

# Demos

## Demo1: total distance and avg speed

compute the total distance and average speed of each trajectory.

```python
summary = []

for traj in traj_collection.trajectories:
    # Duration in seconds
    duration = (traj.get_end_time() - traj.get_start_time()).total_seconds()

    # Total distance (in meters)
    length = traj.get_length()

    # Average speed (m/s)
    avg_speed = length / duration if duration > 0 else 0

    summary.append({
        "trajectory_id": traj.id,
        "total_distance(m)": length,
        "avg_speed(m/s)": avg_speed,
    })

summary_df = pd.DataFrame(summary)
print(summary_df)
```

## Demo3: Mark all the turning point on plot

Mark all the turning point (direction change > 45 degrees & move < 2 meters within 5 tracked point)

```python
fig, ax = plt.subplots(figsize=(10, 8))

TURN_ANGLE_THRESHOLD = 45  # degrees
MOVE_DISTANCE_THRESHOLD = 2  # meters
SHIFT = 5  # number of points to shift forward/backward

for traj in traj_collection.trajectories:
    traj.plot(ax=ax, linewidth=1.5, alpha=0.6, label=f"Traj {traj.id}")

    df = traj.df.copy()

    # Ensure direction exists
    if 'direction' not in df.columns:
        traj.add_direction(overwrite=True)
        df = traj.df.copy()

    # Shift directions forward and backward
    df['dir_prev'] = df['direction'].shift(SHIFT)
    df['dir_next'] = df['direction'].shift(-SHIFT)

    # Compute direction change (angle difference)
    df['angle_change'] = abs(df['dir_next'] - df['dir_prev'])
    df['angle_change'] = df['angle_change'].apply(lambda x: 360 - x if x > 180 else x)

    # Compute displacement over the shift
    df['move_distance'] = df.geometry.distance(df.geometry.shift(SHIFT))

    # Condition: large direction change, small movement
    mask = (df['angle_change'] > TURN_ANGLE_THRESHOLD) & (df['move_distance'] < MOVE_DISTANCE_THRESHOLD)
    turning_points = df[mask].geometry

    # Plot turning points
    turning_points.plot(ax=ax, color='green', markersize=30, label=f"Turning {traj.id}")

ax.legend(loc='upper right')
plt.title('Turning Points (Δdirection > 45° & move < 2m within 5 points)')
plt.xlabel("Easting (m)")
plt.ylabel("Northing (m)")
plt.grid(True)
plt.show()
```
