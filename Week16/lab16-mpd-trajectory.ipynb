## Reading CSVs

read csv into
1. wamvs: dictionary of GeoDataFrame
2. trajs: dictionary of Trajectory

```python
boats = ['wamv1', 'wamv2', 'wamv3', 'wamv4', 'wamv5']
wamvs = {}
trajs = {}

for boat in boats:
    df = pd.read_csv('data/bory_boat/' + boat + '_gps.csv', delimiter=',')
    df['t'] = pd.to_datetime(df['t'], unit='ns')
    wamv = df[['X', 'Y', 'trajectory_id', 't']]
    wamv['geometry'] = df.apply(lambda row: Point(row['X'], row['Y']), axis=1)

    gdf = gpd.GeoDataFrame(wamv, geometry='geometry', crs='EPSG:4326')
    gdf = gdf.to_crs(epsg=32649) # Convert to UTM zone 49N for Taiwan

    trajs[boat] = mpd.Trajectory(gdf, 'trajectory_id', t='t')
    wamvs[boat] = gdf
```

## Demo3: Mark all the turning point on plot

Mark all the turning point (direction change > 45 degrees & move < 2 meters within 5 tracked point)

```python
fig, ax = plt.subplots(figsize=(10, 8))
turning_points = {}
TURN_ANGLE_THRESHOLD = 45
MOVE_DISTANCE_THRESHOLD = 2
SHIFT = 5

for traj in traj_collection.trajectories:
    # First plot the original trajectory
    traj.plot(ax=ax, linewidth=2, alpha=0.7, label=traj.id)
    
    # Get the trajectory's GeoDataFrame
    gdf = traj.df.copy()
    
    # Calculate direction changes and movement distances
    turning_points_list = []
    
    # Iterate through points with a window of SHIFT points
    for i in range(len(gdf) - SHIFT):
        current_point = gdf.iloc[i]
        future_point = gdf.iloc[i + SHIFT]
        
        # Calculate direction change
        direction_change = abs(future_point['direction'] - current_point['direction'])
        # Normalize to 0-180 range
        direction_change = min(direction_change, 360 - direction_change)
        
        # Calculate movement distance
        movement_distance = current_point['geometry'].distance(future_point['geometry'])
        
        # Check if it meets turning point criteria
        if (direction_change > TURN_ANGLE_THRESHOLD and 
            movement_distance < MOVE_DISTANCE_THRESHOLD):
            turning_points_list.append(current_point['geometry'])
    
    # Convert to GeoSeries and plot
    if turning_points_list:
        turning_geoseries = gpd.GeoSeries(turning_points_list)
        turning_geoseries.plot(ax=ax, color='red', markersize=50, marker='x', label=f'Turning points ({traj.id})')
        turning_points[traj.id] = turning_geoseries

ax.legend()
plt.title('Trajectories with turning points (direction change > 45 degrees)')
plt.xlabel("Easting (m)")
plt.ylabel("Northing (m)")
plt.show()
```
